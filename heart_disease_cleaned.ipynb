```python
#importing our dependencies
import pandas as pd
```

```python
#loading in our csv file with the data
df = pd.read_csv('Resources/heart_disease.csv')
df.head()
```

```python
#checking the name of the columns
df.columns
```

```python
na_counts = df.isna().sum()

print(na_counts)
```

```python
df.drop(['education', 'cigsPerDay','BPMeds','glucose'], axis=1, inplace=True)

df.head()
```

```python
# Rename the columns using a mapping dictionary
new_column_names = {
    'Gender': 'gender',
    'age': 'age',
    'currentSmoker': 'current_smoker',
    'prevalentStroke': 'prevalent_stroke',
    'prevalentHyp': 'prevalent_hypertension',
    'diabetes': 'diabetes',
    'totChol': 'total_cholesterol',
    'sysBP': 'systolic_blood_pressure',
    'diaBP': 'diastolic_blood_pressure',
    'BMI': 'bmi',
    'heartRate': 'heart_rate',
    'Heart_ stroke': 'heart_stroke'
}

df.rename(columns=new_column_names, inplace=True)

df.head()
```

```python
#checking the size of the dataframe
df.shape
```

```python
df.dropna(inplace=True)
```

```python
df.shape
```

```python
# Save the cleaned DataFrame to a new CSV file
output_file_path = 'Resources/cleaned_data.csv'
df.to_csv(output_file_path, index=False)

print("DataFrame has been saved to:", output_file_path)
```
